  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**

    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)
- [
- A geodatabase is the storage container to organize spatial data like a folder. It holds multiple types of geographic information in one area.
- In a geodatabase, a feature dataset is like another folder within the geodatabase. The feature dataset groups related data layers together. This helps with organization and consistency.
- Feature classes are individual data layers that are kept inside the feature dataset. The feature classes represent specific geographic elements like points, lines, and polygons. For example, this includes roads and lakes.
- Together, the geodatabase, feature dataset, and feature classes help make GIS work more organized and improves overall workflow. ]
